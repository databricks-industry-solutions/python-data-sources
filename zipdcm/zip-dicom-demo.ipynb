# Read Zipped DICOM files saving time and storage
WIth the custom "zipdcm" Python Data Source, we can read zipped (and non Zipped) up DICOM files directly to extract their metadata.

Requirements:
- Recommend DBR 17.0 (Apache Spark 4.0) compute
- Shared cluster compute compatible
- Working on serverless compute fix.
- Requires `pydicom==3.0.1`

```python
%sh ls -lh ./resources/dcms
```

```python
from dbx.zip_dcm_ds import ZipDCMDataSource
spark.dataSource.register(ZipDCMDataSource)

# read DCMs with `numPartitions` parallelism.
df = spark.read.format("zipdcm").option('numPartitions',4).load("./resources")
df.display()
```
